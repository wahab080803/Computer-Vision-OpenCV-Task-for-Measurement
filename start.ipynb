```python
import cv2
import numpy as np

# Load an image
# Replace 'path/to/your/image.jpg' with the actual path to your image
image = cv2.imread('.\\images\\10.jpg') 

if image is None:
    print("Error: Could not load image. Please check the path.")
else:
    # Add some salt-and-pepper noise for demonstration purposes
    # This part is optional and just to illustrate the effectiveness of median blur
    noise_probability = 0.05
    noisy_image = np.copy(image)
    num_salt = np.ceil(noise_probability * image.size * 0.5)
    coords = [np.random.randint(0, i - 1, int(num_salt)) for i in image.shape]
    noisy_image[coords[0], coords[1], :] = 255  # Salt noise (white)

    num_pepper = np.ceil(noise_probability * image.size * 0.5)
    coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in image.shape]
    noisy_image[coords[0], coords[1], :] = 0  # Pepper noise (black)

    # Apply median blur
    # The second argument is the kernel size, which must be an odd, positive integer.
    # A common starting value is 5.
    median_blurred_image = cv2.medianBlur(noisy_image, 5)

    # Display the original, noisy, and median-blurred images
    cv2.imshow('Original Image', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    cv2.imshow('Noisy Image', noisy_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    cv2.imshow('Median Blurred Image', median_blurred_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    # Wait for a key press and then close all windows
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the image in grayscale
image = cv2.imread('.\\images\\10.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image was loaded successfully
if image is None:
    print("Error: Could not load image.")
    exit()

# Apply Sobel operator in x and y directions
# ddepth=cv2.CV_64F is used to handle negative gradient values
sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3) # Horizontal edges
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3) # Vertical edges

# Combine the results to get the final edge-detected image
# We take the absolute value and then convert to uint8 for visualization
abs_sobelx = cv2.convertScaleAbs(sobelx)
abs_sobely = cv2.convertScaleAbs(sobely)
sobel_combined = cv2.addWeighted(abs_sobelx, 0.5, abs_sobely, 0.5, 0)

# Display the original and edge-detected images
cv2.imshow('Original Image', image)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow('Sobel X', abs_sobelx)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow('Sobel Y', abs_sobely)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow('Sobel Combined', sobel_combined)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

# Load the image
# Ensure 'your_image.jpg' is in the same directory or provide the full path
image = cv2.imread('.\\images\\10.jpg') 

# Check if the image was loaded successfully
if image is None:
    print("Error: Could not load image. Please check the file path.")
else:
    # Convert the image to grayscale (Canny works best on grayscale images)
    gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # Apply Gaussian blur to reduce noise (optional but recommended)
    # The kernel size (5,5) can be adjusted
    blurred_img = cv2.GaussianBlur(gray_img, (5, 5), 0)

    # Apply Canny edge detection
    # The two threshold values (e.g., 100 and 200) are crucial for edge detection.
    # Edges with intensity gradient above the upper threshold are considered strong edges.
    # Edges between the lower and upper thresholds are considered weak edges and are included
    # only if they are connected to strong edges.
    # Edges below the lower threshold are discarded.
    edges = cv2.Canny(blurred_img, 100, 200)

    # Display the original and edge-detected images
    cv2.imshow('Original Image', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    cv2.imshow('Canny Edges', edges)
    # Wait for a key press and then close all windows
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

img = cv2.imread(".\\images\\10.jpg")
cv2.imshow("Original Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()

img = cv2.resize(img, (400, 400))
cv2.imshow("Resized Image", img)
cv2.waitKey(0)
cv2.destroyAllWindows()

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Gray Image", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()

_, thresh = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY_INV)
cv2.imshow("Thresholded Image", thresh)
cv2.waitKey(0)
cv2.destroyAllWindows()

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

print("Number of contours:", len(contours))
# for i, contour in enumerate(contours):
#     cv2.drawContours(img, contours, i, (0, 0, 0), 2)

cv2.drawContours(img, contours, -1, (0, 255, 0), 2)

cv2.imshow("Contours", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
for i, contour in enumerate(contours):
    area = cv2.contourArea(contours[i])
    print(f"Area of contour {i}: {area}")

    perim = cv2.arcLength(contours[i], True)
    print(f"Perimeter of contour {i}: {perim}")
    
    x, y, w, h = cv2.boundingRect(contours[i])
    cv2.rectangle(img, (x, y), (x+w, y+h), (255,0,0), 2)

    rect = cv2.minAreaRect(contours[i])
    
    box = cv2.boxPoints(rect)
    box = np.int0(box)
    cv2.drawContours(img, [box], 0, (0,0,255), 2)

    (x, y), radius = cv2.minEnclosingCircle(contours[i])
    cv2.circle(img, (int(x), int(y)), int(radius), (0,255,255), 2)

cv2.imshow("Contours with Shapes", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# ----------------- CONTOUR AND MEASUREMENT LOGIC -----------------

import math


thresh = img_thresh.copy()
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 1. Define the Known Reference Coin
# Assuming Coin 35 (Area 5920.5) is the 18mm diameter coin
REFERENCE_AREA_PIXELS = 5920.5  # Area of the reference coin (Coin 35)
REFERENCE_DIAMETER_MM = 18.0    # Known real-world diameter in millimeters

# 2. Calculate the Conversion Factor (Cf)
# We need to convert the known diameter to area in mm^2 to find the ratio.
# Area_mm^2 = pi * (Diameter_mm / 2)^2
REFERENCE_AREA_MM2 = math.pi * (REFERENCE_DIAMETER_MM / 2)**2

# Conversion Factor (Cf): How many pixels^2 equal one mm^2
# Cf = Area_pixels^2 / Area_mm^2
PIXELS_PER_MM2 = REFERENCE_AREA_PIXELS / REFERENCE_AREA_MM2
# To convert back to linear unit, we need the square root of Cf
PIXELS_PER_MM = math.sqrt(PIXELS_PER_MM2) 
# Note: For simplicity and robustness, it's often better to just use minEnclosingCircle to get the diameter directly.

print(f"Calculated Conversion Factor (pixels/mm): {PIXELS_PER_MM:.2f}")

coins_data = []  # Store measurement data

for i, cnt in enumerate(contours):
    area = cv2.contourArea(cnt)
    
    # Filter out noise (using a generous threshold now that we have a reference)
    if area < 700:
        continue

    # Use minEnclosingCircle for better diameter measurement of circular objects
    (x_c, y_c), radius = cv2.minEnclosingCircle(cnt)
    D_pixels = 2 * radius  # Diameter in pixels
    
    # Calculate Real-World Diameter
    D_mm = D_pixels / PIXELS_PER_MM
    
    # Get Bounding Box for Drawing
    x, y, w, h = cv2.boundingRect(cnt)
    
    coins_data.append({
        "ID": i,
        "Area": area,
        "Diameter_px": D_pixels,
        "Diameter_mm": D_mm,
        "Rect": (x, y, w, h)
    })
    
    # Draw results on the image
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    cv2.putText(img, f"{D_mm:.1f} mm", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)


cv2.imshow("Detected and Measured Coins", img)
cv2.waitKey(0)
cv2.destroyAllWindows()

# ----------------- FINAL REPORT -----------------
print("\n--- Final Measurement Report ---")
for coin in coins_data:
    print(f"Coin {coin['ID']} (Area: {coin['Area']:.1f}): Diameter = {coin['Diameter_mm']:.1f} mm")

print(f"\nTotal coins successfully measured: {len(coins_data)}")
```
